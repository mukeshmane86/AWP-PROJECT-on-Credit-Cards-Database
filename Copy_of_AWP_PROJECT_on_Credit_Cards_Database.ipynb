```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Pie Chart

first i have created pie charts of data base using "Geography" also given % count using autopct command.the details of code used for to plot graph is as follow.

```python
plt.figure(figsize=(15,7))
df['Geography'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Pie Chart of Geography",fontsize=16)
```

BAR CHART

Next  i have created Bar charts of data base using "Creditscore Vs Coutrywise" the details of code used for to plot graph is as follow.

```python
plt.figure(figsize = (10,7))
sns.barplot(data = df, x ='Geography' , y = 'CreditScore')
plt.title("Creditscore Vs Coutrywise",fontsize=16)
```

HISTIGRAPH CHART

 Next  i have created Histograph  charts of data base using "Creditscore "the details of code used for to plot graph is as follow.

```python
plt.figure(figsize = (20,7))
plt.xticks(rotation=45)
plt.hist(df['CreditScore'])
plt.show()
```

BAR CHART

Next  i have created Bar charts of data base using "Geography Vs Balance" the details of code used for to plot graph is as follow.

```python
x= df['Geography']
y= df['Balance']
plt.xlabel('Geography',fontsize=10)
plt.ylabel('Balance',fontsize=20)
plt.title("Geography Vs Balance",fontsize=16)
plt.xticks(rotation=90)
sns.barplot(x,y)
plt.show()
```

Scatter *plot* for first 500 rows

Next  i have created scatterplot charts of data base using "data=df1,x="CreditScore",y="Age",hue="Geography" as the is huge so i used only first 500 rows data. the details of code used for to plot graph is as follow

```python
df1=df.head(500)
plt.figure(figsize=(10,10))
plt.title("CreditScore Vs Age",fontsize=16)
sns.scatterplot(data=df1,x="CreditScore",y="Age",hue="Geography")
plt.plot()
```

Heat map

Next i have created Heatmap charts  using "Database- Credit cards details " 
the details of code used for to plot graph is as follow

```python
plt.figure(figsize = (15,5))

sns.heatmap(data = df.corr(),annot = True)
```

BAR Chart

Next  i have created Bar charts of data base using "Geography Vs NumOfProducts" the details of code used for to plot graph is as follow

```python
df1=df.head(500)
x= df1['Geography']
y= df1['NumOfProducts']
plt.bar(x,y)
plt.xlabel('Geography',fontsize=10)
plt.ylabel('NumOfProducts',fontsize=20)
plt.title("Geography Vs NumOfProducts",fontsize=16)
plt.xticks(rotation=90)
plt.show()
```
